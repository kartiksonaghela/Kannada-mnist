```python
import numpy as np
import tensorflow as tf
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
#dig_mnist = pd.read_csv("Dig-MNIST.csv")
#sample_submission = pd.read_csv("sample_submission.csv")
train.dtypes
```

```python
x_train=train.drop(['label'], axis = 1)
y_train=train["label"].values
#x_test=dig_mnist.drop(['label'], axis = 1)
#y_test=dig_mnist["label"].values
print(x_train.shape)
```

```python
x_train=(x_train.values/255)
x_train = x_train.reshape((-1,784))
```

```python
x_train=x_train.reshape(60000, 28, 28, 1)
#training_images=training_images / 255.0
#x_test = x_test.reshape(10240, 28, 28, 1)
#test_images=test_images/255.0
x_train.shape
```

```python
def build_model(hp):  
    model = keras.Sequential([
    keras.layers.Conv2D(
        filters=hp.Int('conv_1_filter', min_value=32, max_value=128, step=16),
        kernel_size=hp.Choice('conv_1_kernel', values = [3,5]),
        activation='relu',
        input_shape=(28,28,1)
    ),
    keras.layers.Conv2D(
        filters=hp.Int('conv_2_filter', min_value=32, max_value=64, step=16),
        kernel_size=hp.Choice('conv_2_kernel', values = [3,5]),
        activation='relu'
    ),
    keras.layers.Flatten(),
    keras.layers.Dense(
        units=hp.Int('dense_1_units', min_value=32, max_value=128, step=16),
        activation='relu'
    ),
    keras.layers.Dense(10, activation='softmax')
  ])
  
    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3])),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
  
    return model
```

```python
from kerastuner import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
from tensorflow import keras
```

```python
tuner_search=RandomSearch(build_model,
                          objective='val_accuracy',
                          max_trials=5)
```

```python
tuner_search.search(x_train,y_train,epochs=3,validation_split=0.1)
```

```python
model=tuner_search.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
model.fit(x_train,y_train, epochs=10, validation_split=0.1, initial_epoch=3)
```

```python
test1 = pd.read_csv("test.csv")
test1=test1.drop(['id'], axis = 1)
x_train.shape
```

```python
test1 = (test1/255)
test1 = test1.values.reshape((-1,28,28,1))
test1.shape
```

```python
result = model.predict(test1)
result = np.argmax(result,axis = 1)
result = pd.Series(result,name="Label")
```

```python
submission = pd.concat([pd.Series(range(1,5001),name = "ImageId"),result],axis = 1)
submission.to_csv("digit_recognizer_data_augmentation.csv",index=False)
```
